```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Replacing blanks with 0 as tenure is 0 and no total charges are recorded**

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

**Replacing 0 and 1 to Yes or No for Senior citizen**

```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(30)
```

```python
ax = sns.countplot(x = "Churn", data = df)

ax.bar_label(ax.containers[0])
plt.title("COUNT OF CUSTOMERS BY CHURN")
plt.show()
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb["Churn"], labels = gb.index, autopct = "%1.2f%%")
plt.title("PERCENTAGE OF CHURNED CUSTOMERS")
plt.show()
```

```python
plt.figure(figsize = (5,4))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("GENDER BASED ON CHURN")
plt.show()
```

```python
plt.figure(figsize = (5,4))
sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
plt.title("CHURN BY SENIOR CITIZEN")
plt.show()
```

```python
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert to percentages
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Step 3: Plot
plt.figure(figsize=(5, 5))

bottom = [0] * len(ct_percent)
colors = ['#1f77b4', '#ff7f0e']  # Blue, Orange (standard matplotlib palette)

for i, col in enumerate(ct_percent.columns):
    plt.bar(ct_percent.index, ct_percent[col], 
            bottom=bottom, label=col, color=colors[i])
    
    # Add percentage labels
    for j, val in enumerate(ct_percent[col]):
        plt.text(j, bottom[j] + val/2, f'{val:.1f}%', ha='center', va='center', fontsize=9)
    
    bottom = [bottom[k] + ct_percent[col].iloc[k] for k in range(len(ct_percent))]

# Final touches
plt.title("CHURN BY SENIOR CITIZEN (PERCENT STACKED)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.xticks([0, 1], ['No', 'Yes'])  # 0 = No, 1 = Yes
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Define orange and blue color palette
custom_palette = ['#ff7f0e', '#1f77b4']  # Orange and Blue

# Set up subplot grid
n_cols = 3
n_rows = -(-len(columns) // n_cols)  # Ceiling division

plt.figure(figsize=(15, 12))

for i, col in enumerate(columns, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(x=col, data=df, palette=custom_palette, hue = "Churn")
    plt.title(col)
    plt.xticks(rotation=45)
    plt.xlabel("")
    plt.ylabel("Count")

plt.tight_layout()
plt.show()
```
